```python
data = pd.read_csv('dataset_thumbnail_featured.csv', sep=";")
data.head()
```

```python
# Handling missing values by filling null description and tags with empty strings
data['desc'] = data['desc'].fillna("")
data['tags'] = data['tags'].fillna("")
```

```python
# Converting upload_date column to datetime objects for time based calculations
data["upload_date"]= pd.to_datetime(data["upload_date"])
# Using the latest upload date in the dataset as a reference point for time calculations
reference_date = data["upload_date"].max()
data["days_since_upload"] = (reference_date - data["upload_date"]).dt.days

# --- TIME FEATURES ---
# Extracting month and year to capture seasonal and yearly trends
data['upload_month'] = data['upload_date'].dt.month
data['upload_year'] = data['upload_date'].dt.year

# Extracting specific time units like hour and day of the week
data["publish_hour"] = data["upload_date"].dt.hour
data["publish_day_of_week"] = data["upload_date"].dt.day_of_week
# Creating a binary flag to identify if a video was posted during the weekend
data["is_weekend"] = data["publish_day_of_week"].apply(lambda x:1 if x>=5 else 0 )

def categorize_hour(hour):
    #Categorizes the 24 hour clock into descriptive time periods to help the model understand the impact of posting times.
    if 0 <= hour < 6:
        return 'Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'PrimeTime'  
    else:
        return 'LateNight'
# Applying the categorization logic to create a new categorical feature
data['time_of_day'] = data['publish_hour'].apply(categorize_hour)

# Using One Hot Encoding to convert categorical time periods into numerical columns
data = pd.get_dummies(data, columns=['time_of_day'], prefix='time_part', drop_first=False)

```

```python
# TEXT BASED FEATURES
# Calculating basic metrics for video titles like length and word count
data["title_length"] = data["video_title"].apply(len)
data["title_word_count"] = data["video_title"].apply(lambda x: len(x.split()))
# Counting uppercase characters which often indicates clickbait or shouting
data["title_uppercase_count"] = data["video_title"].apply(lambda x: sum(1 for c in x if c.isupper()))
# Checking for special characters like question marks or exclamation points
data["has_special_char"] = data["video_title"].apply(lambda x: 1 if '?' or'!' in x else 0)
# Calculating title complexity based on average character count per word
data['title_complexity'] = data['title_length'] / (data['title_word_count'] + 1)

    # Shout factor: Combined ratio of uppercase letters and special characters (clickbait indicator)
data['shout_factor'] = (data['title_uppercase_count'] + data['has_special_char']) / (data['title_length'] + 1)
```

```python
# DESCRIPTION AND TAG FEATURES
# Counting how many tags are used per video
data["tag_count"] = data['tags'].apply(lambda x: len(x.split(',')) if x.strip() != "" else 0)
# Measuring the length of the video description
data["desc_length"] = data["desc"].apply(len)
# Counting external links found in the description
data["link_count"] = data["desc"].apply(lambda x: x.lower().count("http"))

# Identifying social media mentions in the description
social_medias= ["facebook","twitter","instagram","tiktok","linkedin","twitch","discord"]
def count_social_media(text):
    text = text.lower()
    return sum(1 for platform in social_medias if platform in text)
data["social_media_count"] = data["desc"].apply(count_social_media)
```

```python
# SENTIMENT ANALYSIS
from textblob import TextBlob


def get_sentiment(text):
    #Uses TextBlob to calculate the emotional tone (polarity) of the title. Value ranges from -1 (Negative) to +1 (Positive).
    try:
        return TextBlob(text).sentiment.polarity
    except:
        return 0

data['title_sentiment'] = data['video_title'].apply(get_sentiment)
```

```python
# VIRALITY AND REACH POTENTIAL
    # Interaction between follower count and time since upload (Logarithmic scale)
    # This helps the model understand the theoretical "upper limit" of views
data['reach_potential'] = np.log1p(data['follower_count'] * data['days_since_upload'])
data['follower_log'] = np.log1p(data['follower_count'])
```

```python
# VISUAL ATTRACTIVENESS (THUMBNAIL)
# Visual Pop calculates the interaction between color saturation and perceived brightness
data['visual_pop'] = (data['thumb_saturation'] * data['thumb_perceived_brightness']) / 100

# Color Variety measures the standard deviation across RGB channels to determine color diversity
data['color_variety'] = np.std(data[['thumb_r', 'thumb_g', 'thumb_b']], axis=1)
```

```python
cols_to_drop= ["","",""]
```

```python
# Dropping irrelevant or target related columns that should not be used in the training process
columns_to_drop = ["query_used","search_type","like_count","comment_count","video_url"]
data.drop(columns= [c for c in columns_to_drop if c in data.columns ] ,inplace=True )
```
